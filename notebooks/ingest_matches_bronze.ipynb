```python
import json 
from pyspark.sql.functions import col, to_timestamp

JSON_PATH = '/Volumes/workspace/sports-pipeline/sports-pipeline/matches_20260413_212419.json'

df_raw = spark.read.option("multiline", "True").json(JSON_PATH)
df_raw.printSchema()
df_raw.show(5, truncate=False)
```

```python
df_matches = df_raw.select(
    col("id").cast("string").alias("match_id"),
    col("utcDate").alias("match_date"),
    col("status"),
    col("homeTeam.name").alias("home_team"),
    col("awayTeam.name").alias("away_team"),
    col("score.fullTime.home").alias("home_score"),
    col("score.fullTime.away").alias("away_score"),
    col("competition.name").alias("competition")
)

df_matches.show(10)
```

```python
#Save Delta Table
df_matches.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("`sports-pipeline`.football_matches_delta")

print("Delta Table created!")
```

```python
#Save Iceberg
df_matches.write \
    .format("iceberg") \
    .mode("overwrite") \
    .saveAsTable("`sports-pipeline`.football_matches_iceberg")

print("Iceberg Table created!")
```
